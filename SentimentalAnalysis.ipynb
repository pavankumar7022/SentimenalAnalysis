```python
# --------------------------------------------
# Sentiment Analysis with TextBlob + pandas
# --------------------------------------------

# Install required libraries
!pip install textblob pandas

# Import necessary packages
from textblob import TextBlob
import pandas as pd

# --------------------------------------------
# Step 1: Enter or import your sentences
# --------------------------------------------
texts =[
    "I love Python programming!",
    "This movie was terrible and boring.",
    "I am feeling okay today.",
    "The food was amazing!",
    "I donâ€™t like rainy days.",
    "The product is good but a bit expensive."
]
    
print("sentence:",texts)

# --------------------------------------------
# Step 2: Analyze sentiment for each sentence
# --------------------------------------------
results = []

for sentence in texts:
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    # Determine sentiment category
    if polarity > 0:
        sentiment = "Positive ğŸ˜€"
    elif polarity < 0:
        sentiment = "Negative ğŸ˜"
    else:
        sentiment = "Neutral ğŸ˜"

    # Append results
    results.append({
        "Sentence": sentence,
        "Polarity": polarity,
        "Subjectivity": subjectivity,
        "Sentiment": sentiment
    })

# --------------------------------------------
# Step 3: Create a DataFrame for display
# --------------------------------------------
df = pd.DataFrame(results)

# Show the results table
print("texts:")
print("âœ… Sentiment Analysis Results:\n")
display(df)

# --------------------------------------------
# Step 4: Save results to a CSV file
# --------------------------------------------
csv_filename = "sentiment_results.csv"
df.to_csv(csv_filename, index=False)

print(f"\nğŸ’¾ Results have been saved to '{csv_filename}' in your current directory.")
```
